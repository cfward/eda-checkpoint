```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Specify the file name "Anime.csv"
csv_file = "Anime.csv"

# Read the CSV file into a DataFrame
df = pd.read_csv(csv_file)

# Display the first few rows of the DataFrame to get a glimpse of the data
print(df.head())
```

## Get to Know the Numbers

Now that you have everything setup, put any code that you use to get to know the dataframe and its rows and columns better in the cell below. You can use whatever techniques you like, except for visualizations. You will put those in a separate section.

When working on your code, make sure to leave comments so that your mentors can understand your thought process.

```python
# Get the dimensions of the DataFrame (number of rows and columns)
print("Shape of the DataFrame:", df.shape)

# List the column names in the DataFrame
print("Column Names:", df.columns)

# Get basic statistics for numerical columns (e.g., count, mean, std, min, max)
print("Summary Statistics for Numerical Columns:")
print(df.describe())

# Check data types and non-null values for each column
print("Data Types and Missing Values:")
print(df.info())


# Get unique values in the 'Type' column
unique_types = df['Type'].unique()
print("Unique Values in 'Type' column:", unique_types)

# Count the occurrences of each unique value in the 'Type' column
type_counts = df['Type'].value_counts()
print("Value Counts for 'Type' column:")
print(type_counts)
```

## Visualize

Create any visualizations for your EDA here. Make note in the form of code comments of what your thought process is for your visualizations.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for Seaborn plots
sns.set(style="whitegrid")

# Create a histogram for the 'Score' column to visualize its distribution
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Score', bins=30, kde=True, color='hotpink')
plt.title('Distribution of Anime Scores')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.show()

# Create a bar plot to visualize the distribution of anime types
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Type', palette='RdPu')
plt.title('Distribution of Anime Types')
plt.xlabel('Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
